**Importieren von benötigten Bibliotheken sowie lesen der CSV-Datei**

```python
import pandas as pd

filepath = 'Verkaufszahlen_Kaggle.csv' # Verkaufszahlen für Usecase 2
df = pd.read_csv(filepath, decimal='.')
```

**Ausgabe des Dataframes**

```python
df
```

**Entfernen der nicht benötigten Spalten**

```python
new_df = df.drop(columns=['Invoice ID', 'Branch', 'City', 'Customer type', 'Product line', 'Tax 5%', 'Date', 'Time', 'Payment', 'cogs', 'gross margin percentage', 'gross income', 'Quantity', 'Unit price' ])
new_df.head()
```

**Gesamtausgaben (Total) aufrunden**

```python
new_df.round({'Total': 2})
```
